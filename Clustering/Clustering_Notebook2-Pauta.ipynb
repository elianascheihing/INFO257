(a) Mclust es un método que permite ajustar un modelo de mezcla de normales a un conjunto de datos, bajo el supuesto que las variables involucradas se distribuyen según normales multivariadas. El modelo es estimado usando el algoritmo EM, tomando como agrupamiento inicial el obtenido mediante un agrupamiento jerárquico aglomerativo. El modelo óptimo es obtenido utilizando el criterio de información Bayesiano (BIC).

Sus principales parámetros de entrada son:

- data: los datos de las variables que serviran para hacer el clustering
- G: Nro de Grupos, por omisión el método evalua desde 1, 2 a 9 grupos.
- modelNames: string describiendo las distintas configuraciones de la distribución normal multivariada utilizada en el algoritmo EM. Por omisión el método evalúa 14 configuraciones distintas
- prior: posible distribución a priori definida por el usuario, por omisión no se provee.
- control: parámetros de control del algoritmo EM, por omisión no se provee.
- initialization: parámetros para el clustering inicial, por omisión no se provee.
- warn: valor lógico asociado al manejo de advertencias (warnings)
- x: un objeto de la clase mclustBIC que utiliza un ajuste anterior para proveer los valores BIC de los distintos modelos.

La salida del método es un objeto de la clase Mclust, cuyas principales componentes son:

- data: la matriz con los datos en la entrada
- modelName: String que describe el modelo que obtuvo el valor óptimo del BIC:
    - "EII": varianza-covarianza esférica de igual volumen.
    - "VII": varianza-covarianza esférica volumen variable.
    - "EEI": varianza-covarianza diagonal de igual volumen y forma.
    - "VEI": varianza-covarianza diagonal, volumen variable, de igual forma. 
    - "EVI": varianza-covarianza diagonal, de igual volumen y  forma variable.
    - "VVI": varianza-covarianza diagonal, de  volumen y  forma variable.
    - "EEE": varianza-covarianza elipsoidal, de igual volumen, forma y orientación
    - "VEE": varianza-covarianza elipsoidal, volumen variable, de igual forma y orientación
    - "EVE": varianza-covarianza elipsoidal, de igual volumen y orientación, forma variable
    - "VVE": varianza-covarianza elipsoidal,  volumen y forma variable e igual orientación
    - "EEV": varianza-covarianza elipsoidal,  de igual volumen y forma variable y orientación variable.
    - "VEV": varianza-covarianza elipsoidal,  volumen y orientación variable e igual forma.
    - "EVV": varianza-covarianza elipsoidal,  orientación y forma variable e igual volumen.
    - "VVV": varianza-covarianza elipsoidal,  volumen, forma y orientación variables 

- n: número de observaciones en los datos
- d: dimensión de los datos (Nro de variables)
- G: el número óptimo de grupos
- BIC: todos los valores de BIC calculados.
- ICL: log de la verosimilitud del modelo con BIC óptimo
- df: Número de parámetros estimados
- bic: valor BIC del modelo seleccionado
- z: una matriz cuya componente [i,k] es la probabilidad de que la observación i en el data set - pertenezca a la clase k.

**Criterio de Información Bayesiano**

Una forma de discriminar entre posibles agrupamientos, es calcular el Criterio de Información Bayesiano, cuya expresión genérica es:
$$BIC = log L(Datos\mid \hat{\theta}) - \frac{d}{2}log(N)$$
donde $d$ es el número de parámetros libres y $N$ el número de observaciones. A medida que aumenta el número de parámetros la verosimilitud aumenta, pero esto es moderado por el término en que depende de $d$.

Mclust calcula automática el BIC para distintas configuraciones de las matrices covarianza de cada Normal multivariada (que corresponden a distintos números de parámetros) así como distinto número de grupos. Por parsimonia, es recomendable escoger el primer número de grupos que representa un máximo local.

**1. Purity (Pureza):** Mide qué tan bien están agrupados los elementos en clusters homogéneos, o entre un clustering  y  un etiquetado en clases pre-existente. Se calcula de la siguiente manera:

$$ Purity  =  \sum_{i=1,\cdots,K} \frac{N_i}{N} p_{i} = \frac{1}{N} \sum_{i=1,\cdots,K} \max_{j=1,\cdots,K}(C_{ij})$$


Donde:

$N$ es el número total de elementos.

$N_i$ es el número de elementos del cluster $i$

$p_{ij}= \frac{N_{ij}}{N_i}, j=1,\cdots,K$ es la distribución empírica de las etiquetas en el cluster $i$

y entonces

$$p_i = \max_{j=1,\cdots,K} p_{ij}$$

La segunda expresión considera de manera equivalente: 

$C_{ij}$ es la cantidad de elementos de la clase $j$ asignados al cluster $i$, con $i,j=1,\cdots,K$ 



Asi, la pureza evalua cuánto se aproximan los clusters a representar unas clases homogéneas conocidas del contexto (etiquetado pre-existente). Un valor de pureza cercano a 1 indica que los elementos están agrupados correctamente en clusters que reflejan fielmente las clases originales.

**2. Rand Index:** 
Mide la similitud entre dos asignaciones de clustering (o una clasificación y un clustering)  al comparar todos los pares de elementos. Se calcula de la siguiente manera:
$$RI =  \frac{(TP + TN)}{(TP + TN + FN + FP)}$$

Donde:

$TP$ es el número de pares de elementos que están en la misma clase y en el mismo cluster, conocidos como "verdaderos positivos"

$TN$ es el número de pares de elementos que están en diferentes clases y en diferentes clusters, conocidos como "verdaderos negativos"

$FN$ es el número de pares de elementos que están en la misma clase pero en diferentes clusters, conocidos como "falsos negativos".

$FP$ es el número de pares de elementos que están en diferentes clases pero en el mismo cluster, conocidos como "falsos positivos"

Al igual que Purity,  Rand Index compara una asignación en clases (pre-existente) con el clustering desarrollado para evaluar cuánto coinciden en términos de las relaciones entre los elementos. Un valor de $RI$ cercano a 1 indica una alta similitud entre las asignaciones del clustering y la clasificación pre-existente, lo que significa que los elementos están agrupados de manera similar en ambos casos.


**3. Mutual Information (Información Mutua):**
La información mutua mide la cantidad de información común entre dos asignaciones de clustering o un clustering ($U= \cup U_i$) y una clasificación pre-existente ($V= \cup V_j$). Se calcula de la siguiente manera:

$$MI = \sum_{i=1,\cdots,K} \sum_{j=1,\cdots,K} p_{ij} \log_2\left ( \frac{p_{ij}}{p_i p_j}\right )$$



Donde:

$$p_{ij} =  \frac{|U_i \cap V_j|}{N}$$

es la probabilidad conjunta de que un elemento pertenezca al cluster i en $U$ y a la clase j en $V$.

Y por otra parte se definen: 
$$p_i =\frac{|U_i|}{N} \mbox{  y   } p_j = \frac{|V_j|}{N} $$ 

como  la probabilidad de que un elemento pertenezca al cluster i en $U$
y la probabilidad de que un elemento pertenezca a la clase j en $V$, respectivamente. 

La información mutua mide la cantidad de información compartida entre dos asignaciones de clustering. Un valor más alto de información mutua indica una mayor similitud y consistencia entre las dos asignaciones de clustering. Una alta información mutua significa que los clusters están capturando de manera efectiva las relaciones entre los elementos y las clases.
